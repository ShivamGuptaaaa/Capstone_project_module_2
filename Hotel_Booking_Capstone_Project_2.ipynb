#**EDA**

#Overview of canceled bookings

```python
# Count the number of 'is_canceled' value
is_canceled_count = df['is_canceled'].value_counts()
is_canceled_count
```

```python
# Plot the graph for total number of bookings VS canceled or non-canceled
plt.rcParams['figure.figsize'] = (10, 5)
is_canceled_count.plot(kind='barh')
plt.title('Non-canceled & Canceled booking graph')
plt.xlabel('Total number of bookings')
plt.ylabel('Booking Cancelled (No = 0, Yes = 1)')

# 0 means not cancel, 1 means cancel
```

*   With the help of Non-canceled & Canceled booking graph, we can see that approx 75000 of customers checked-in to the hotel while approx 44000 of customers canceled their bookings.

```python
# Finding the percentage of bookings
booking_percentage = is_canceled_count/is_canceled_count.sum()*100
booking_percentage
```

```python
# Plot the graph for non-canceled VS canceled booking percentage
plt.figure(figsize=(6,4))
booking_percentage.plot(kind = 'bar', color='blue')
plt.title('Non-canceled vs canceled booking percentage graph')
plt.xlabel('Booking Canceled (No = 0, Yes = 1)')
plt.ylabel('Booking percentage')
plt.grid()
```

*  From the Non-canceled vs canceled booking percentage graph, we can clearly see that around 37% of bookings got canceled whereas around 63% customer did Check-in.

#Most Busiest month for hotel (Peak month for hotels)

```python
# Percentage distribution of month arrivals
df['arrival_date_month'].value_counts(normalize=True)*100
```

```python
# Plot the count plot for arrivals per month
plt.figure(figsize=(12,8) ,dpi=110)
sns.countplot(x= 'arrival_date_month', data = df, order = df['arrival_date_month'].value_counts().index,palette='Blues_r')
plt.title("Arrival Per Month", weight = 'bold')
plt.xlabel("Arrival Month")
plt.ylabel("Count")
```

*  The above Arrival Per Month graph to represent the number of booking each month

#  Month having highest and lowest booking rates


```python
# Group the arrival according to the month and finding the mean of ADR
df.groupby(['arrival_date_month','hotel'])['adr'].mean().unstack()
```

```python
# Plot the line graph between ADR(average Daily Rate) and Arrival month
plt.figure(figsize=(12,6), dpi=110)
sns.lineplot(x='arrival_date_month', y='adr', data=df, hue='hotel')
plt.title("Arrival Perv Month", weight='bold')
plt.legend(title='Hotel Type')
plt.xlabel('Arrival Month')
plt.ylabel('ADR (Average Daily Rate)')
plt.grid()
```

*   This lineplot clearly shows that prices in the Resort Hotel are much higher during the August month.

*   This lineplot shows that prices in City Hotel varies less than Resort Hotel and is more expensive in May and June month.

#  Most booked accommodation type - Single, Couple or Family




```python
# Select single, couple, multiple adults and family
df_not_canceled = df[df['is_canceled']==0]
```

```python
# Select single, couple, multiple adults and family
single = df_not_canceled[(df_not_canceled.adults == 1) & (df_not_canceled.children == 0) & (df_not_canceled.babies == 0)]
couple = df_not_canceled[(df_not_canceled.adults == 2) & (df_not_canceled.children == 0) & (df_not_canceled.babies == 0)]
family = df_not_canceled[df_not_canceled.adults + df_not_canceled.children + df_not_canceled.babies > 2 ]
```

```python
# Make the list of Catergory names and their total percentage
names = ['Single','Couple (No children)','Family/Friends']
count = [single.shape[0],couple.shape[0],family.shape[0]]
count_percenatge = [x/df_not_canceled.shape[0]*100 for x in count]

# Create a dictionary to make a dataframe
d = {'names':names, 'count_percent':count_percenatge}
acc_type = pd.DataFrame(data = d)

# See DataFrame
acc_type
```

```python
# Plot the graph between accommodation type and their percentage of customer
color=['#1f77b4', '#ff7f0e', '#2ca02c']
plt.figure(figsize=(10,7), dpi=110)
sns.barplot(data = acc_type, x = 'names', y = 'count_percent', palette=color)
plt.title('Accommodation Type', weight = 'bold')
plt.xlabel('Type of customer')
plt.ylabel('Booking percentage')
```

*   From barplot, it is seen that most number of customer are containing 2 adults customer.
*   And those who have more than 2 either containing adults, children & babies have the lowest number of customer.

#  Bookings are more on Weekdays or Weekends


```python
# Calculate total bookings for weekdays and weekends
total_weekend_bookings = df['stays_in_weekend_nights'].sum()
total_weekday_bookings = df['stays_in_week_nights'].sum()
```

```python
# Prepare pie chart
labels = ['Weekends', 'Weekdays']
sizes = [total_weekend_bookings, total_weekday_bookings]
colors = ['#ff7f0e', '#1f77b4']

# Create a pie chart
plt.figure(figsize=(6, 6), dpi=110)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Set the title
plt.title('Bookings on Weekdays vs Weekends', weight='bold')

plt.axis('equal')
plt.show()
```

*  As we see in the Bookings on Weekdays vs Weekends, 72.9% booking done on weekdays and only 27.1% booking done on weekdays.
*  More booking done on Weekdays.
